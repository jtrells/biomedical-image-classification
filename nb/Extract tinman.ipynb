```python
subfigures_cursor = db.figures.find({'state': 'Reviewed', 'type': 'Subfigure'})
processed_subfigures = []
for subfigure in subfigures_cursor:
    figure = db.figures.find_one({"_id": subfigure['figureId']})
    subfigure['caption'] = figure['caption']
    
    mods = list(set([mapping(x, modalities, modalities_ids) for x in subfigure['modalities']]))
    number_modalities = len(mods)
    
    if 'composition' in subfigure:
        composition = subfigure['composition']
    else:
        composition = ''

    new_subfigure = {
        'img': str(subfigure['_id']),
        'source': 'tinman',
        'img_path': '/tinman' + subfigure['uri'],
        'caption': figure['caption'],
        'split_set': 'TRAIN',
        'modality': ';'.join(mods),
        'flag': subfigure['flag'] if 'flag' in subfigure else False,
        'needsCropping': subfigure['needsCropping'],
        'isCompound': subfigure['isCompound'],
        'isOvercropped': subfigure['isOvercropped'],
        'isMissingSubfigures': subfigure['isMissingSubfigures'],
        'isMissingPanels': subfigure['isMissingPanels'],
        'numberSubpanes': subfigure['numberSubpanes'],
        'username': subfigure['username'],
        'observations': subfigure['observations'],
        'isOverfragmented': subfigure['isOverfragmented'] if 'isOverfragmented' in subfigure else False,
        'composition': composition,
        'number_modalities': number_modalities
    }
    processed_subfigures.append(new_subfigure)
```

```python
import pandas as pd
df = pd.DataFrame(processed_subfigures)
df.head()
```

```python
df_train = df[df.flag==False]
```

```python
def mapping(mod_id, mod_vals, mod_idx):
    _id = mod_id['_id']
    x = mod_vals[mod_idx.index(str(_id))]
    
    if x[1] == 'Experimental':
        if x[2] == 'Microscopy':            
            label = f'MICROSCOPY,{x[3].upper()}'
            if x[4] == 'Transmission EM':
                label += ',TRANSMISSION'
            elif x[4] == 'Scanning EM':
                label += ',SCANNING'
            elif 'Other' in x[4]:
                label += ',OTHER'
            else:
                label += f',{x[4].upper()}'
        elif x[2] == 'Gel':
            label = f'EXPERIMENTAL,GEL,{x[3].upper()}'
        elif x[2] == 'Plate':
            label = f'EXPERIMENTAL,PLATE'
    elif x[1] == 'Molecular Structure':
        if x[2] == 'Macromolecule Sequence':
            label = f'MOLECULAR STRUCTURE,{x[3].upper()}'
        else:            
            label = f'MOLECULAR STRUCTURE,{x[2].upper()}'
    elif x[1] == 'Graphics':
        if x[2] == 'Bar Chart':
            label = 'GRAPHICS,HISTOGRAM'
        elif x[2] == 'Line Chart':
            label = 'GRAPHICS,LINECHART'
        elif x[2] == 'Graph':
            label = 'GRAPHICS,FLOWCHART'
        elif x[2] == 'Heatmap':
            label = 'GRAPHICS,OTHER'
        else:
            label = f'GRAPHICS,{x[2].upper()}'
    elif x[1] == 'Organs and Organisms':
        if x[2] == 'X-Ray':
            label = 'RADIOLOGY,X-RAY'
        elif x[2] == 'MRI and CT':
            label = 'RADIOLOGY,CT/MRI/PET'
        elif x[2] == 'Photographs':            
            label = 'PHOTOGRAPHY' # lucky there are no others
    elif x[1] == 'Other':
        label = 'OTHER'
    else:
        label = ''
    return label
    
            
        
    
```
