```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
DATA_PATH = '../data/ActiveLearning/'
```

```python
experiments_to_compare = os.listdir('../data/ActiveLearning/higher_modality/500/1e-05/')
experiments_to_compare = [i for i in experiments_to_compare if '.ipynb' not in i]
experiments_to_compare = [i for i in experiments_to_compare if 'class_weights_True' in i]
experiments_to_compare = [i for i in experiments_to_compare if 'DELTA_0.005' not in i]
experiments_to_compare = [i for i in experiments_to_compare if 'DELTA_0.02' not in i]
#experiments_to_compare = [i for i in experiments_to_compare if ('CEAL_True' in i)|('rs' in i)]
experiments_to_compare
```

```python
metrics = ['Accuracy','Macro F1-Score','Macro Precision','Macro Recall']
```

```python
df_concat =pd.DataFrame()
for experiment in experiments_to_compare:
    path = os.path.join(DATA_PATH,'higher_modality/500/1e-05',experiment,'metrics.csv')
    df = pd.read_csv(path,sep = '\t').iloc[:-2,:]
    df['Experiment'] = df['Sample Method'].astype(str) +'_' + df['CEAL'].astype(str)
    df_concat = pd.concat([df_concat,df],axis = 0).reset_index(drop = True)

for metric in metrics:
    sns.lineplot(data=df_concat, x="Labeled (%)", y=metric, hue="Experiment")
    plt.show()
```

```python
filter_var = 'lc'
def func_string(x):
    if 'False' in x:
        return x.replace('_False','')
    if 'True' in x:
        return 'CEAL_'+ x.replace('_True','')  


df = df_concat[(df_concat['Sample Method']==filter_var)|(df_concat['Sample Method']=='rs')].reset_index(drop = True)
df['Macro-Recall (%)'] = df['Macro Recall']
df['Accuracy (%)'] = df['Accuracy']
df['Experiment']   = df['Experiment'].apply(lambda x:func_string(x))
#sns.lineplot(data=df, x="Labeled (%)", y='Macro-Recall (%)', hue="Experiment",style="Experiment")
#plt.show()
color_ = 'm'
for i in df.Experiment.unique():
    if 'CEAL' in i:
        df_ = df[df['Experiment']==i].reset_index(drop = True)
        plt.plot(df_['Labeled (%)'].values, df_['Macro-Recall (%)'], color=color_,label =df_['Experiment'].values[0])
    elif 'rs' in i:
        df_ = df[df['Experiment']==i].reset_index(drop = True)
        plt.plot(df_['Labeled (%)'].values, df_['Macro-Recall (%)'],'--', color='k',label = df_['Experiment'].values[0])
    else:
        df_ = df[df['Experiment']==i].reset_index(drop = True)
        plt.plot(df_['Labeled (%)'].values, df_['Macro-Recall (%)'], '--',color=color_,label = df_['Experiment'].values[0])
    
plt.legend(title = 'Method',fancybox = True)
plt.xlabel('Labeled (%)')
plt.ylabel('Macro-Recall (%)')
plt.show()
```
