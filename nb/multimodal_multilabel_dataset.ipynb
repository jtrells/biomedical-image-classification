```python
import pandas as pd
csv_path = '/workspace/data/multimodality_classification.csv'

df = pd.read_csv(csv_path, sep='\t')
df.head()
```

```python
df.shape
```

```python
multi_df = df.pivot(index='ID', columns='MODALITY', values='MODALITY').reset_index()
multi_df.head()
```

```python
def get_path(img_id): return df.loc[df['ID'] == img_id]['PATH'].values[0]
def get_caption(img_id): return df.loc[df['ID'] == img_id]['CAPTION'].values[0]
def get_set(img_id): return 'TEST' if df.loc[df['ID'] == img_id]['SET'].values[0] == 'TEST' else 'TRAIN'
```

```python
get_path('11373_2007_9226_Fig1_HTML-10.jpg')
```

```python
get_caption('11373_2007_9226_Fig1_HTML-10.jpg')
```

```python
get_set('11373_2007_9226_Fig1_HTML-10.jpg')
```

```python
multi_df['CAPTION'] = multi_df.apply(lambda x: get_caption(x['ID']), axis=1)
multi_df['PATH'] = multi_df.apply(lambda x: get_path(x['ID']), axis=1)
multi_df['SET'] = multi_df.apply(lambda x: get_set(x['ID']), axis=1)
multi_df = multi_df.fillna(0)
multi_df['DMEL'] = multi_df.apply(lambda x: 1 if x['DMEL'] != 0 else 0, axis=1)
multi_df['DMFL'] = multi_df.apply(lambda x: 1 if x['DMFL'] != 0 else 0, axis=1)
multi_df['DMLI'] = multi_df.apply(lambda x: 1 if x['DMLI'] != 0 else 0, axis=1)
multi_df['DMTR'] = multi_df.apply(lambda x: 1 if x['DMTR'] != 0 else 0, axis=1)
```

```python
import numpy as np

multi_df = multi_df.rename(columns={"ID": "IMAGE_NAME"})
multi_df['ID'] = np.arange(len(multi_df))
```
